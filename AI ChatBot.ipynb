```python
from datetime import datetime
import random
import time
```

```python
# ===========================
# RULE-BASED AI CHATBOT
# ===========================

print("=" * 60)
print("🤖      WELCOME TO RULE-BASED AI CHATBOT")
print("=" * 60)
```

```python
name = input("Enter your name: ").strip().title()

print(f"\nHello {name}! 👋")
print("I'm your Rule-Based AI Assistant.")
print("Type 'help' to see available commands.")
print("Type 'bye' anytime to exit.\n")
```

```python
conversation_count = 0
history = []
```

```python
while True:

    user_message = input(f"{name}: ").strip().lower()
    conversation_count += 1

    history.append(f"{name}: {user_message}")

    print("Bot is typing...")
    time.sleep(1)

    # ===========================
    # Greeting
    # ===========================

    if user_message in ["hello", "hi", "hey", "good morning", "good afternoon", "good evening"]:

        response = f"Hello {name}! 😊 How can I help you today?"

    # ===========================
    # How are you
    # ===========================

    elif user_message == "how are you":

        response = "I'm doing great! Thanks for asking. 😊"

    # ===========================
    # Who are you
    # ===========================

    elif user_message == "who are you":

        response = "I am a Rule-Based AI Chatbot created using Python."

    # ===========================
    # AI
    # ===========================

    elif user_message == "what is ai":

        response = (
            "Artificial Intelligence (AI) enables machines "
            "to perform tasks that normally require human intelligence."
        )

    # ===========================
    # Thanks
    # ===========================

    elif user_message in ["thanks", "thank you"]:

        response = "You're welcome! 😊"

    # ===========================
    # Time
    # ===========================

    elif user_message == "time":

        current_time = datetime.now().strftime("%I:%M %p")
        response = f"Current time is {current_time}"

    # ===========================
    # Date
    # ===========================

    elif user_message == "date":

        current_date = datetime.now().strftime("%d-%m-%Y")
        response = f"Today's date is {current_date}"

    # ===========================
    # Motivation
    # ===========================

    elif user_message == "motivate me":

        quotes = [
            "Success comes to those who never give up.",
            "Every expert was once a beginner.",
            "Believe in yourself and keep learning.",
            "Small daily improvements lead to big results.",
            "Your future is created by what you do today."
        ]

        response = random.choice(quotes)

    # ===========================
    # Help
    # ===========================

    elif user_message == "help":

        print("\n========== AVAILABLE COMMANDS ==========")
        print("👋 hello / hi / hey")
        print("🌅 good morning")
        print("🌇 good afternoon")
        print("🌃 good evening")
        print("😊 how are you")
        print("🤖 who are you")
        print("🧠 what is ai")
        print("🙏 thanks")
        print("🕒 time")
        print("📅 date")
        print("💪 motivate me")
        print("❓ help")
        print("👋 bye")
        print("========================================\n")
        continue

    # ===========================
    # Exit
    # ===========================

    elif user_message == "bye":

        print("\nBot: Thank you for chatting with me. 😊")
        print(f"Bot: You asked {conversation_count} question(s).")

        print("\nConversation History")

        for chat in history:
            print(chat)

        print("\nBot: Goodbye! Have a wonderful day. 👋")
        break

    # ===========================
    # Unknown Command
    # ===========================

    else:

        response = (
            "Sorry, I don't understand that command.\n"
            "Type 'help' to see all available commands."
        )

    history.append("Bot: " + response)

    print("Bot:", response)
```
