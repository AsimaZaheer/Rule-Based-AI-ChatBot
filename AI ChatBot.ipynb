```python
from datetime import datetime
import random


print("=" * 50)
print("🤖 Welcome to Rule-Based AI Chatbot")
print("Type 'bye' anytime to exit.")
print("=" * 50)

while True:

    user_message = input("You: ").strip().lower()

    if user_message in ["hello", "hi", "hey"]:
        print("Bot: Hello! 👋 How can I help you today?")

    elif user_message == "how are you":
        print("Bot: I am doing great! Thanks for asking. 😊")

    elif user_message == "who are you":
        print("Bot: I am a Rule-Based AI Chatbot created using Python.")

    elif user_message == "what is ai":
        print("Bot: AI stands for Artificial Intelligence. It enables machines to perform tasks that normally require human intelligence.")

    elif user_message == "thanks":
        print("Bot: You're welcome! 😊")

    elif user_message == "thank you":
        print("Bot: My pleasure! Happy to help. 😊")

    elif user_message == "time":
        current_time = datetime.now().strftime("%I:%M %p")
        print(f"Bot: Current time is {current_time}")

    elif user_message == "date":
        current_date = datetime.now().strftime("%d-%m-%Y")
        print(f"Bot: Today's date is {current_date}")

    elif user_message == "motivate me":

        quotes = [
            "Success comes to those who never give up.",
            "Every expert was once a beginner.",
            "Believe in yourself and keep learning.",
            "Small daily improvements lead to big results.",
            "Your future is created by what you do today."
        ]

        print("Bot:", random.choice(quotes))

    elif user_message == "help":
      print("\n===== AVAILABLE COMMANDS =====")
      print("👋 hello")
      print("👋 hi")
      print("👋 hey")
      print("😊 how are you")
      print("🤖 who are you")
      print("🧠 what is ai")
      print("🙏 thanks")
      print("🙏 thank you")
      print("🕒 time")
      print("📅 date")
      print("💪 motivate me")
      print("❓ help")
      print("👋 bye")
      print("==============================")

    elif user_message == "bye":
        print("Bot: Goodbye! Have a nice day. 😊")
        break

    else:
        print("Bot: Sorry, I don't understand.")
```
